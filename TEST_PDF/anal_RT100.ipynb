```python
from numpy import *
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
ener_NP0800 = loadtxt('NP0800_LD20P3_C025_RT100_PDF/NP0800_LD20P3_C025_RT100_PDF.ener')
ener_NP1600 = loadtxt('NP1600_LD20P3_C025_RT100_PDF/NP1600_LD20P3_C025_RT100_PDF.ener')
ener_NP2400 = loadtxt('NP2400_LD20P3_C025_RT100_PDF/NP2400_LD20P3_C025_RT100_PDF.ener')
```

```python
Nc = 10.
C_rep = 25.
Rt = 100.
t_NP0800 = ener_NP0800[:,0]/(C_rep*Rt)
t_NP1600 = ener_NP1600[:,0]/(C_rep*Rt)
t_NP2400 = ener_NP2400[:,0]/(C_rep*Rt)
fNAS_NP0800 = 100.*ener_NP0800[:,4]/(Nc*800.)
fNAS_NP1600 = 100.*ener_NP1600[:,4]/(Nc*1600.)
fNAS_NP2400 = 100.*ener_NP2400[:,4]/(Nc*2400.)
```

```python
plt.figure(figsize=(11,4))
plt.plot(t_NP0800, fNAS_NP0800, 'b-', label = 'NP0800')
plt.plot(t_NP1600, fNAS_NP1600, 'g-', label = 'NP1600')
plt.plot(t_NP2400, fNAS_NP2400, 'r-', label = 'NP2400')
plt.legend(loc = 'lower right')
plt.xlabel(r'time / $\tau_0$', fontsize=20)
plt.ylabel(r'fNAS (%)', fontsize=20)
```
